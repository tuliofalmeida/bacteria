# Concatenating the files

First let's store the path in a variable. After cloning the repository, the entire repository will be downloaded to this virtual notebook. You can check all files using the menu on the left

```python
path_clists = '/content/bacteria/data_tutorial'
```

This process will take according to the number of FOVs (clists) and the size of these FOVs (frames)

```python
df2d,df3d,_ = bac.concatenate_clist(path_clists, direct = True) # we use the '_' at the end of the func because we don't need the last output
```

After this step, our data is stored in dataframes

```python
#@title 2D data
df2d.head()
```

```python
#@title 3D data
df3d.head()
```

## Filters

These data are already pre-filtered using the parameter 'stat0==2' estimated by SuperSegger. If you don't want it, you can pass 'filter=False' in the concatenate_clist() function. Here you can see the effect of the filters on the data and choose which one you want to use (daughter=True, lentgh=True).

_stat0: Indicator of cell birth and death. stat0 = 0 means no birth was observed, stat0 =1 means no death/division was observed , stat0 = 2 means both birth and death were observed._

```python
df2d_f,df3d_f,_ = bac.combined_filters(df2d,df3d,daughter=True, lentgh=True, plot=True)
```

```python
#@title 2D data filtered
df2d_f.head()
```

```python
#@title 3D data filtered
df3d_f.head()
```
